```
def PatternCount(Text,Pattern):
  count = 0
  for i in range (len(Text) - len(Pattern)+1):
    if Pattern == Text[i:i+len(Pattern)]:
      count +=1
  return count


def FrequencyMap(Text, k):
    freq={}
    n = len(Text)
    for i in range(n-k+1):
        Pattern = Text[i:i+k]
        freq[Pattern] = PatternCount(Text,Pattern)
    
    # for i in range(n-k+1):
    #     Pattern = Text[i:i+k]
    #     if Pattern in freq:
    #       freq[Pattern] +=1              
   
    return freq

def FrequentWords(Text, k):
    words = []
    freq = FrequencyMap(Text, k)
    m = max(freq.values())
    for key in freq:
      if freq[key] == m:
        words.append(key)
        
    return words

print(FrequentWords("CGATATATCCATAG",3))
```

```
### Reverse 
def Reverse(Pattern):
  reverse = ""
  for char in Pattern:
    reverse = char + reverse
  
  return reverse


# def Complement(Pattern):
#   complement = ""
#   for char in Pattern:
#     if char == "A":
#       complement = complement + "T"
#     elif char == "C":
#       complement = complement + "G"
#     elif char == "G":
#       complement = complement + "C"
#     else:
#       complement = complement + "A"
#   return complement

def Complement(Pattern):
    basepairs = {"A":"T", "G":"C", "T":"A", "C":"G"}
    complement = ""
    for base in Pattern:
        complement += basepairs.get(base)
    return complement

def ReverseComplement(Pattern):
  reverse = Reverse(Pattern)
  complement = Complement(reverse)

  return complement

Pattern = "AAAACCCGGT"
ReverseComplement(Pattern)


```

```
def PatternMatching(Pattern,Genome):
  positions = []
  for base in range(len(Genome) - len(Pattern)+1):
    if(Pattern == Genome[base:base+len(Pattern)]):
      positions.append(base)
  return positions

Pattern = "ATAT"
Genome = "GATATATGCATATACTT"

PatternMatching(Pattern,Genome)
```
